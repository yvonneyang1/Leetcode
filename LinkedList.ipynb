## Leetcode No.2    Add Two Numbers
You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. 
Add the two numbers and return the sum as a linked list.
You may assume the two numbers do not contain any leading zero, except the number 0 itself.
## Leetcode No.445    Add Two Numbers II
The most significant digit comes first and each of their nodes contain a single digit.

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        head = ListNode(0)
        sumList = head
        carry = 0
        while l1 or l2:
            if l1 and l2:
                val, carry = (l1.val + l2.val + carry) % 10, (l1.val + l2.val + carry) // 10
                sumList.next = ListNode(val)
                l1 = l1.next
                l2 = l2.next
                sumList = sumList.next       
            elif not l1:
                val, carry = (l2.val + carry) % 10, (l2.val + carry) // 10
                sumList.next = ListNode(val)
                l2 = l2.next
                sumList = sumList.next
            else:
                val, carry = (l1.val + carry) % 10, (l1.val + carry) // 10
                sumList.next = ListNode(val)
                l1 = l1.next
                sumList = sumList.next
        if carry > 0:
            sumList.next = ListNode(carry)   
                
        return head.next   
    
#  Problem 2
#  Method: 1. Reverse input
#  Follow up: Don't reverse input
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        # find the length of both lists
        n1 = n2 = 0
        curr1, curr2 = l1, l2
        while curr1:
            curr1 = curr1.next 
            n1 += 1
        while curr2:
            curr2 = curr2.next 
            n2 += 1
            
        # parse both lists and sum the corresponding positions 
        # without taking carry into account
        curr1, curr2 = l1, l2
        head = None
        while n1 > 0 and n2 > 0:
            val = 0
            if n1 >= n2:
                val += curr1.val 
                curr1 = curr1.next 
                n1 -= 1
            if n1 < n2:
                val += curr2.val 
                curr2 = curr2.next
                n2 -= 1     
            # update the result: add to front
            curr = ListNode(val)
            curr.next = head
            head = curr

        # take the carry into account
        # to have all elements to be less than 10
        curr1, head = head, None
        carry = 0
        while curr1:
            # current sum and carry
            val = (curr1.val + carry) % 10
            carry = (curr1.val + carry) // 10
            
            # update the result: add to front
            curr = ListNode(val)
            curr.next = head
            head = curr

            # move to the next elements in the list
            curr1 = curr1.next
        
        # add the last carry
        if carry:
            curr = ListNode(carry)
            curr.next = head
            head = curr

        return head
    
```

## Leetcode No.206    Reverse Linked List

```python
class Solution:
    def reverseList(self, head: ListNode) -> ListNode:       
        prev, curr = None, head
        
        while curr:
            Next = curr.next
            curr.next = prev
            prev = curr
            curr = Next
        return prev
```

## Leetcode No.21    Merge Two Sorted Lists

```python
class Solution:
    def mergeTwoLists(self, l1: ListNode, l2: ListNode) -> ListNode:
        dumy = ListNode(0)
        head = dumy
        while l1 and l2:
            if l1.val < l2.val:
                head.next = l1
                l1 = l1.next
                head = head.next
            else:
                head.next = l2
                l2 = l2.next
                head = head.next
        head.next = l1 if l1 else l2        
        return dumy.next                  
```

## Leetcode No.138    Copy List with Random Pointer

A linked list is given such that each node contains an additional random pointer which could point to any node in the list or null.
### Return a deep copy of the list.
Each node is represented as a pair of [val, random_index] where:
val: an integer representing Node.val
random_index: the index of the node (range from 0 to n-1) where random pointer points to, or null if it does not point to any node.

```python
class Solution:
    def copyRandomList(self, head: 'Node') -> 'Node':
        self.nodeMap = {}
        
        def helper(node):
            if not node:
                return None
            if node in self.nodeMap:
                return self.nodeMap[node]
            new_node = Node(node.val, None, None)
            self.nodeMap[node] = new_node
            new_node.next = helper(node.next)
            new_node.random = helper(node.random)
            return new_node
       
        return helper(head)
```

## Leetcode No.234    Palindrome Linked List

```python
class Solution:
    def isPalindrome(self, head):
        rev = None
        slow = fast = head
        while fast and fast.next:
            fast = fast.next.next
            rev, rev.next, slow = slow, rev, slow.next
        if fast:
            slow = slow.next
        while rev and rev.val == slow.val:
            slow = slow.next
            rev = rev.next
        return not rev
```

## Leetcode No.92    Reverse Linked List II
Reverse a linked list from position m to n. Do it in one-pass.
Note: 1 ≤ m ≤ n ≤ length of list.

```python
def reverseBetween(self, head, m, n):
        dummy = ListNode(0)
        dummy.next = head
        
        cur, prev = head, dummy
        for _ in range(m - 1):
            cur = cur.next
            prev = prev.next
        
        for _ in range(n - m):
            temp = cur.next
            cur.next = temp.next
            temp.next = prev.next
            prev.next = temp

        return dummy.next
```

## Leetcode No.203    Remove Linked List Elements

```python
def removeElements(self, head: ListNode, val: int) -> ListNode:
        if not head:
            return 
        if not val:
            return head
        dummy = ListNode()
        dummy.next = head
        prev, curr = dummy, head
        while curr:
            if curr.val == val:
                curr = curr.next
                prev.next = curr
            else:
                prev = curr
                curr = curr.next
        return dummy.next
```

### Method
1. fast, slow points find the middle node
2. reverse the second part 
3. merge

## Leetcode No.86 Partition List
Given the head of a linked list and a value x, partition it such that all nodes less than x come before nodes greater than or equal to x.
You should preserve the original relative order of the nodes in each of the two partitions.

```python
def partition(self, head: ListNode, x: int) -> ListNode:
        left , right = ListNode(0), ListNode(0)
        left_head, right_head = left, right
        while head:
            if head.val < x:
                left.next = head
                left = left.next
            else:
                right.next = head
                right = right.next
            head = head.next
            
        left.next = right_head.next
        right.next = None
        
        return left_head.next
```

## Leetcode No.148    Sort List
Follow up: sort the linked list in O(n logn) time and O(1) memory

```python
def sortList(self, head: ListNode) -> ListNode:
        def split(node):
            if not node or not node.next:
                return node
            left, fast, slow = node, node, node
            fast = fast.next
            while fast and fast.next:
                fast = fast.next.next
                slow = slow.next
            right = slow.next
            slow.next = None
            left_sort = split(left)
            right_sort = split(right)
            return merge(left_sort, right_sort)       
        
        def merge(l, r):           
            dummy = ListNode(0)
            tail = dummy
            while l and r:
                if l.val <= r.val:
                    tail.next = l
                    l = l.next
                else:
                    tail.next = r
                    r = r.next
                tail = tail.next
            tail.next = r if r else l
            return dummy.next
        
        return split(head)
```
